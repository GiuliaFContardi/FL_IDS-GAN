In summary, you should prioritize analyzing high correlation relationships between features in your scatter plots to understand feature importance, detect patterns, and assess GAN performance. However, don't ignore random relationships, as they may reveal additional insights into the data generation process.

```python
alphaTrasparenza = 0.4
```

```python
images_dir = '/content/drive/MyDrive/Colab Notebooks/progettoTesi/grafici_pdf/vanillaGan/setHBase/Botnet_tanti'
#images_dir = '/content/drive/MyDrive/Colab Notebooks/progettoTesi/grafici_pdf/vanillaGan/setHBuono/Botnet_tanti'
```

```python
fig, axs = plt.subplots(1, 3, figsize=(20, 6))


axs[0].scatter(df.iloc[:, 6], df.iloc[:, 13], label='Dati Reali')
axs[0].scatter(gen_samples.iloc[:, 6], gen_samples.iloc[:, 13], label='Dati Generati', c='orange', alpha=alphaTrasparenza)
axs[0].set_title("PacketLengthStd & PacketLengthMean") #corr =0.94
axs[0].legend()


axs[1].scatter(df.iloc[:, 9], df.iloc[:, 21], label='Dati Reali')
axs[1].scatter(gen_samples.iloc[:, 9], gen_samples.iloc[:, 21], label='Dati Generati', c='orange', alpha=alphaTrasparenza)
axs[1].set_title("RSTFlagCount & FwdRSTFlags") #corr =0.90
axs[1].legend()

axs[2].scatter(df.iloc[:, 18], df.iloc[:, 23], label='Dati Reali')
axs[2].scatter(gen_samples.iloc[:, 18], gen_samples.iloc[:, 23], label='Dati Generati', c='orange', alpha=alphaTrasparenza)
axs[2].set_title("AveragePacketSize & SubflowBwdBytes") #corr =0.99
axs[2].legend()

#plt.tight_layout()
plt.suptitle("Scatter plot di alcune relazioni tra feature con alta correlazione >=0.90", fontsize=15);

plt.savefig(f"{images_dir}/VanillaGan_scatterPlot_setHBase_Botnet_10000_CORR.pdf", dpi=150, format='pdf')

plt.show()
```

```python
fig, axs = plt.subplots(1, 3, figsize=(20, 6))

axs[0].scatter(df.iloc[:, 27], df.iloc[:, 32], label='Dati Reali')
axs[0].scatter(gen_samples.iloc[:, 27], gen_samples.iloc[:, 32], label='Dati Generati', c='orange', alpha=alphaTrasparenza)
axs[0].set_title("FwdIATMax & FwdIATTotal") #correlazione alta con 0.66
axs[0].legend()

axs[1].scatter(df.iloc[:, 0], df.iloc[:, 32], label='Dati Reali')
axs[1].scatter(gen_samples.iloc[:, 0], gen_samples.iloc[:, 32], label='Dati Generati', c='orange', alpha=alphaTrasparenza)
axs[1].set_title("FlowDuration & FwdIATTotal") #correlazione alta con 1.00
axs[1].legend()

axs[2].scatter(df.iloc[:, 12], df.iloc[:, 26], label='Dati Reali')
axs[2].scatter(gen_samples.iloc[:, 12], gen_samples.iloc[:, 26], label='Dati Generati', c='orange', alpha=alphaTrasparenza)
axs[2].set_title("BwdPacketLengthMean & BwdInitWinBytes") #correlazione bassa = 0.00
axs[2].legend()

#plt.tight_layout()
plt.suptitle("Scatter plot di alcune relazioni interessanti tra feature (casuali)", fontsize=15);

plt.savefig(f"{images_dir}/VanillaGan_scatterPlot_setHBase_Botnet_10000_CASUAL.pdf", dpi=100, format='pdf')

plt.show()
```

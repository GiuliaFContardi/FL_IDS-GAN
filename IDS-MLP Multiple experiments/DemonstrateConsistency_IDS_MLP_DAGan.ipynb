```python
def calculate_summary_statistics(metrics):
    mean = np.mean(metrics)
    std = np.std(metrics)
    min_val = np.min(metrics)
    max_val = np.max(metrics)
    return mean, std, min_val, max_val


metrics_dict = {
    "Accuracy": accuracies,
    "Balanced Accuracy": balanced_accuracies,
    "F1 Score": f1_scores,
    "Macro F1 Score": macro_f1_scores,
    "Precision": precisions,
    "Macro Precision": macro_precisions,
    "Recall": recalls,
    "Macro Recall": macro_recalls
}


for metric_name, metric_values in metrics_dict.items():
    mean, std, min_val, max_val = calculate_summary_statistics(metric_values)
    print(f"{metric_name}: Mean = {mean:.4f}, Std = {std:.4f}, Min = {min_val:.4f}, Max = {max_val:.4f}")
```
